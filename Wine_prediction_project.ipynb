```python
def correlation(dataset, threshold):
    col_matrix = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                matrix = corr_matrix.columns[i]
                col_matrix.add(matrix)
    return col_matrix
```

```python
final_matrix = correlation(X_train, 0.7)
```

```python
len(set(final_matrix))
```

```python
X_train.drop(final_matrix, axis=1, inplace=True)
X_test.drop(final_matrix, axis=1, inplace=True)
```
